<h3>Traffic Signs</h3>
Traffic signs are all over Pittsburgh. There are many types of signs, and each type can tell us different details about each neighborhood.<br>
<br>
Signs can also tell us about driving quality in each neighborhood. The easier the drive is, the better the neighborhood.
<h4>Wayfinding Signs</h4>
Wayfinding signs are signs with street names and arrows that point you in the right direction. The more wayfinding signs there are, the more streets and intersections there are in the neighborhood.<br><br>
Here are the neighborhoods with the most signs

```python
import pandas
import matplotlib

# Read in the dataset and sort the values
signs = pandas.read_csv('https://data.wprdc.org/datastore/dump/d078a6b5-83a3-4723-a3a9-5371cfe1cc0c')
signs.sort_values('neighborhood')

# Filter function
def getOnlyWayfindingSigns(x):
    return x.description.iloc[0] == "Wayfinding Signage"

# Get all neighborhoods
neighborhoods = signs.neighborhood.unique()

# Create a new dataframe with the total number of wayfinding signs
wayfinding = pandas.DataFrame(signs.groupby('description').filter(getOnlyWayfindingSigns)['neighborhood'].value_counts(), index=neighborhoods)

# Sort the dataframe, fill the NaN values with zero
wayfinding = wayfinding.sort_values('neighborhood',ascending=False).fillna(value=0)
wayfinding.head(10)
```

<h4>Intersections and Warning Signs</h4>
Intersection signs are signs that warn drivers of intersections, like four way or Y intersections. Warning signs warn drivers of upcoming stop signs, traffic lights, or potential hazards. Stop sign and traffic light ahead signs are usually posted because the driver will not be able to see the object until they are right in front of it.<br>
These signs indicate stops or difficult driving conditions.<br><br>
Here are the neighborhoods with the most signs, and the least signs.

```python
# Filter function
def getOnlyWarningSigns(x):
    return "Intersection Warning" in x.description.iloc[0] or "Ahead" in x.description.iloc[0]

warning = pandas.DataFrame(signs.groupby('description').filter(getOnlyWarningSigns)['neighborhood'].value_counts(), index=neighborhoods)
warning = warning.sort_values('neighborhood',ascending=False).fillna(value=0)
warning.head(10)
```
